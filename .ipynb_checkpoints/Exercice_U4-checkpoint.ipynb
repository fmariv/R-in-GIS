### Ejercicio 1

```r
# Crear polígono
x1 <- c(2, 4, 7, 4, 2)
y1 <- c(8, 2, 11, 7, 8)
xy1 <- list(cbind(x1, y1))
poly1 <- st_polygon(xy1)
class(poly1)
plot(poly1)
```

```r
# Convertir a multipolígono
multipoly <- st_cast(poly1, 'MULTIPOLYGON')
class(multipoly)
plot(multipoly)
```

```r
# Convertir de nuevo a polígono
poly2 <- st_cast(multipoly, 'POLYGON')
class(poly2)
plot(poly2)
```

### Ejercicio 2

```r
# Leer shapefile NC_Counties
nc_counties_path = 'datos/datos_vectoriales/NC_Counties.shp'
nc_counties = st_read(nc_counties_path)
```

### Ejercicio 3

```r
# Carga de archivos
nc_hydro_path = 'datos/datos_vectoriales/NC_Hydro.shp'
nc_hydro <- st_read(nc_hydro_path)
```

```r
# Comprovar validez
st_is_valid(nc_counties)
st_is_valid(nc_webmercator)
```

```r
# Comprovar simplicidad
st_is_simple(nc_counties)
st_is_simple(nc_webmercator)
```

```r
# Crear campo de area
nc_webmercator$Area <- st_area(nc_webmercator) / 1000000
names(nc_webmercator)
```

```r
# Crear campo de longitud
nc_hydro$Longitud <- st_length(nc_hydro) / 1000000
names(nc_hydro)
```

```r
# Crear mapa
plot.new()

plot(st_geometry(nc_counties), axes = TRUE, main = 'Ríos y estados de EUA', xlab = 'Coordenada X', ylab = 'Coordenada Y')
plot(st_geometry(nc_hydro), add = TRUE, col = 'blue')
```
